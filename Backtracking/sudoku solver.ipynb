```python

def sudoku():
    row = 0
    col = 0
    
    #if all cells are assigned than the sudoku is solved
    n = unassigned_no(row,col)
    if n[2] == 0:
        return True
    
    row = n[0]
    col = n[1]
    
    #numberbetween 1-9
    
    for i in range(1,10):
        #if we can assign i to the cell or not
        #the cell is matrix[row][col]
        
        if isSafe(i,row,col):
            matrix[row][col] = i
            
            if sudoku():
                return True
            
            matrix[row][col] = 0
    return False


def unassigned_no(row,col):
    num = 0
    for i in range(0,size):
        for j in range(0,size):
            #cell is un_assigned
            if matrix[i][j] == 0:
                row = i
                col = j
                num = 1
                a = [row,col,num]
                return a
    a = [-1,-1,num]
    return a


def isSafe(n ,row, col):
    for i in range(0,size):
        if matrix[row][i] == n:
            return False
        
    for i in range(0,size):
        if matrix[i][col] == n:
            return False
        
    row_start = (row//3)*3
    col_start = (col//3)*3
    
    for i in range(row_start,row_start+3):
        for j in range(col_start,col_start+3):
            if matrix[i][j] == n:
                return False
            
    return True


matrix = [[ int(ele) for ele in input().split() ]for i in range(9)]
size = 9
ans = sudoku()
if ans is True:
    print('true')
else:
    print('false')
```
